```python
def part_of_img(img,x,x2,y,y2):
    part_of_img = img[y:y2, x:x2]
    return part_of_img
```

```python
def height_up(img,x,y,h): #유지할 가로, 현재 세로, 더할 세로 길이(+)
    height_up = cv2.resize(img,dsize=(x,y+h),interpolation=cv2.INTER_AREA)
    return height_up
```

```python
def height_down(img,x,y,h): #유지할 가로, 현재 세로, 뺄 세로 길이(+)
    height_down = cv2.resize(img,dsize=(x,y-h),interpolation=cv2.INTER_AREA)
    return height_down
```

```python
def width_up(img,x,w,y): #현재 가로, 더할 가로(+), 유지할 세로
    width_up = cv2.resize(img,dsize=(x+w,y),interpolation=cv2.INTER_AREA)
    return width_up
```
